```python
import json

# Load the JSON file
with open('../data/invaders_juliette_22_03_25.json', 'r') as file:
    data = json.load(file)

# Display the loaded data
print(data)
```

```python
import matplotlib.pyplot as plt

# Plot histogram of the values in the data dictionary with one bin per distinct value
plt.hist(data.values(), bins=range(min(data.values()), max(data.values()) + 2), edgecolor='black', align='left')
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
df_cleaned = pd.read_csv('../data/cleaned_invaders.csv')
print(df_cleaned.head())
```

```python
formatted_invaders = [f"PA_{int(invader.split('_')[1]):04d}" for invader in data.keys() if int(invader.split('_')[1]) <= 1600]
```

```python
print(formatted_invaders)   
```

```python
for invader_id in formatted_invaders:
    mask = df_cleaned['id'] == invader_id
    df_cleaned.loc[mask, 'Flashed'] = True
```

```python
print(df_cleaned['Flashed'].sum())
print(len(formatted_invaders))
missing_invaders = [invader_id for invader_id in formatted_invaders if invader_id not in df_cleaned['id'].values]
print(missing_invaders)
```

```python
with open('../data/missing_invaders_flashed.json', 'w') as file:
    json.dump(missing_invaders, file, indent=4)
```

```python
df_cleaned.to_csv('../data/cleaned_invaders.csv', index=False)
```

```python
# Generate the full range of invader numbers
all_invader_numbers = set(range(1, 1538))

# Extract the numeric part of the IDs in df_cleaned
df_cleaned_numbers = set(int(invader.split('_')[1]) for invader in df_cleaned['id'])

# Find the missing numbers
missing_numbers = sorted(all_invader_numbers - df_cleaned_numbers)
print(missing_numbers)
formatted_missing_numbers = [f"PA_{num:04d}" for num in missing_numbers]
print(formatted_missing_numbers)
with open('../data/formatted_missing_invaders_on_map.json', 'w') as file:
    json.dump(formatted_missing_numbers, file, indent=4)
```
